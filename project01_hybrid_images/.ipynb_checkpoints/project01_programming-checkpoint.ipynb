```python
from scipy.ndimage import convolve
from matplotlib.pyplot import imread, imshow
import matplotlib.pyplot as plt
import skimage
import numpy as np
from skimage.transform import rescale
```

```python
img = imread("questions/RISDance.jpg")[1600:2000,2200:2600,:]
imshow(img)
orig_h, orig_w, _ = img.shape
print(img.shape)
```

```python
kh, kw = 9, 9  # Kernel height, width
kernel = np.ones((kh, kw)) / (kh*kw)
padh = (kh) // 2 
padw = (kw) // 2 
img_pad = np.pad(img,[(padh,padh),(padw,padw),(0,0)], mode="constant")
imshow(img_pad)
print(img_pad.shape)
```

```python
imgh, imgw, _ = img_pad.shape 
print(img_pad.shape)
i = 1
result = list()
# Sliding image operation
for hix in range(0,imgh-kh+1,1):
    for wix in range(0,imgw-kw+1, 1):
        y = 0
        #print(hix,hix+kh, wix,wix+kw)
        tmp_img = img_pad[hix:hix+kh, wix:wix+kw]
        #convolution operation
        for kh_ix in range(kh):
            for kw_ix in range(kw):
                y += tmp_img[kh_ix, kw_ix] * kernel[kh-1-kh_ix, kw-1-kw_ix]  #TODO!!!!
        result.append(y)
```
